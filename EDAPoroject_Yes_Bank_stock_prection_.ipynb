## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Convert Date column into datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%b-%y')

# Extract Year and Month
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

# Check final dataset
df.head()
````

### What all manipulations have you done and insights you found?

The Date column was converted into datetime format to enable time-based analysis. Year and Month features were extracted from the Date column for better understanding of stock price trends over time.

The dataset was checked for missing values and duplicate records, and no issues were found. Therefore, no data imputation or duplicate removal was required.

Key insights:

* The dataset contains complete and clean data.
* The Close price is the target variable for prediction.
* Open, High, and Low prices are expected to have a strong relationship with the Close price.
* The extracted Year and Month features can help identify long-term trends and seasonal patterns in stock prices.
